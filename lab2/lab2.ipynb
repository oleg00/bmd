```python
import matplotlib.pyplot as plot
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

2. _Побудова графіків_

```python
plot.plot(x, y)
plot.xlabel('X label')
plot.ylabel('Y label')
plot.title('myfirstplot`stitle')
```

```python
plot.subplot(1, 2, 1)
plot.plot(x, y, 'r')

plot.subplot(1, 2, 2)
plot.plot(y, x, 'b')
```

2.2 Об'єктно-орієнтований метод

```python
figure = plot.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('sample')
```

```python
figure2 = plot.figure()

axes1 = figure2.add_axes([0, 0, 1, 1])
axes2 = figure2.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('xlabel')
axes1.set_ylabel('ylabel')
axes1.set_title('title1')

axes2.plot(y, x, 'b')
axes2.set_xlabel('y label')
axes2.set_ylabel('x label')
axes2.set_title('title2')
```

```python
figure, axes = plot.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(y,x)

# axes.plot(x, y, 'r')
# axes.set_xlabel('x')
# axes.set_ylabel('y')
# axes.set_title('titlesubplot')
```

```python
figu, axes = plot.subplots(nrows=4, ncols=4)
plot.tight_layout
```

```python
figure = plot.figure()

axes = figure.add_axes([0,0,2,2])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend()
```

```python
figure = plot.figure(figsize=(13,37), dpi = 1337)
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, color = 'b')
axes.plot(x, x**3, color = 'g')
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, lw = 3, ls ='--', color='grey')
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, marker = 'o', markersize = 13, markerfacecolor = 'w')
```
