```python
import seaborn as sns
%matplotlib inline
```

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.distplot(dots['firing_rate'])
```

```python
sns.jointplot(x='firing_rate', y='coherence', data=dots, kind='hex')
```

```python
sns.pairplot(dots, palette=sns.color_palette("Paired"), hue='choice')
```

```python
sns.barplot(x='choice', y='firing_rate', data=dots)
```

```python
import numpy as np
sns.barplot(x='choice', y='firing_rate', data=dots, estimator = np.median)
```

```python
sns.boxplot(x="coherence", y="firing_rate", data=dots, palette='rainbow')
```

```python
sns.stripplot(x="coherence", y="firing_rate",data=dots)
```

```python
sns.heatmap(dots.corr())
```
